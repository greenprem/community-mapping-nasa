```python
df = pd.read_csv('final.csv')
```

```python
numeric_df = df.drop(columns=['State'])
```

```python
correlation_matrix = numeric_df.corr()
```

```python
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
```

```python
sns.pairplot(df)
```

```python
sns.lmplot(x='Income', y='Avg_Life_Expectancy', data=df)
plt.title('Income vs Average Life Expectancy')
plt.show()
```

```python
sns.lmplot(x='Literacy', y='Avg_Life_Expectancy', data=df)
plt.title('Literacy vs Average Life Expectancy')
plt.show()
```

```python
sns.lmplot(x='IMR', y='Avg_Life_Expectancy', data=df)
plt.title('Infant Mortality Rate vs Average Life Expectancy')
plt.show()
```
